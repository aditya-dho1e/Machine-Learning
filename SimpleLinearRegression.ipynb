# Simple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Data Preprocessing

```python
dataset = pd.read_csv('D:\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 4 - Simple Linear Regression\Python\Salary_data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Training Simple Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
# regression is predicting continuos values whereas classification is predicting category or class
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

# Predicting the Results

```python
y_pred = regressor.predict(x_test)
```

# Visualising the Results (Training and Test)

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.xlabel('Yrs Exp.')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, regressor.predict(x_test), color='blue')
plt.xlabel('Yrs Exp.')
plt.ylabel('Salary')
plt.show()
```

```python
plt.plot(y_test, label='test')
plt.plot(y_pred, label='pred')
plt.legend()
```
